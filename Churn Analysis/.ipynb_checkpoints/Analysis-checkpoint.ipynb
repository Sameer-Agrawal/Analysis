```python
# A simple way to maintain big data sets is to use CSV files (comma separated files).
# CSV files hold's up plain text and is a well know format that can be read by everyone including Pandas.

# Pandas is a Python library used for working with data sets. It hold's up method's for analyzing, and manipulating data.
import pandas  # Library
import os  # Module

# A module is a set of code or method's having .py extension.

cwd = os.getcwd()  # The os.getcwd() method return the current working / active directory in Python.

# To analyze the datatype of a variable in Python, we can use the built-in type() function.
# An inline function is one for which the compiler copy the code from the function definition directly, instead of creating a separate set of instructions in memory.
# The listdir() method is provided by the os module, and this method, return a list of identifier of all the files provided path.

datum_directory = os.path.join(cwd,os.listdir(cwd)[-1])  # join() method in Python join one or more path components intelligently.

pandas.read_csv(datum_directory + '\\Datum.csv')
```

```python
dataframe = pandas.read_csv(datum_directory + '\\Datum.csv')
dataframe.drop(labels='customerID',axis='columns',inplace=True)  # The drop() method removes the specified row or column
```

```python
# Rudimentary Datum analysis

import numpy

# 1. SHAPE --> shape() method, returns a tuple, that indicates the dimensions of a Python object.
shape = numpy.shape(dataframe)  # Data point represent, a 21 dimensional vector in space
print(shape)
```

```python
# 2. Metadata corresponding attribute --> Return, attribute metadata
metadata = dataframe.info()
print(metadata)

# Observation --> Mutate data type of 'TotalCharges' attribute
```

```python
# Datum mutation

# Datum translation
# The replace() method replaces the specified value with another specified value.

Partner = dataframe['Partner'].replace('Yes',1).replace('No',-1)
Gender = dataframe['gender'].replace('Male',1).replace('Female',-1)
Dependents = dataframe['Dependents'].replace('Yes',1).replace('No',-1)
Phone_Service = dataframe['PhoneService'].replace('Yes',1).replace('No',0)
Multiple_Lines = dataframe['MultipleLines'].replace('No phone service',0).replace('Yes',1).replace('No',-1)
Internet_Service = dataframe['InternetService'].replace('DSL',1).replace('Fiber optic',2)
Online_Security = dataframe['OnlineSecurity'].replace('Yes',1).replace('No',-1)
Device_Protection = dataframe['DeviceProtection'].replace('Yes',1).replace('No',-1)
Tech_Support = dataframe['TechSupport'].replace('Yes',1).replace('No',-1)
Streaming_TV = dataframe['StreamingTV'].replace('Yes',1).replace('No',-1)
Streaming_Movies = dataframe['StreamingMovies'].replace('Yes',1).replace('No',-1)
Contract = dataframe['Contract'].replace('Month-to-month',1).replace('One year',2)
Paperless_Billing = dataframe['PaperlessBilling'].replace('Yes',1).replace('No',-1)
Churn = dataframe['Churn'].replace('Yes',1).replace('No',-1)
```
